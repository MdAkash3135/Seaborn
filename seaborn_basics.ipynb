```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
sns.get_dataset_names()
```

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
penguins['species'].value_counts()
```

```python
sns.set_style("ticks")
sns.scatterplot(data=penguins, x = "flipper_length_mm", y = "body_mass_g", hue="island", palette="Dark2")
sns.despine(left=True)
sns.set_context("paper")
```

```python
sns.scatterplot(data=penguins, x = "species", y = "body_mass_g", hue="island", palette="Dark2", style="sex", alpha = 0.9)
sns.despine(left=True)
sns.set_context("paper")
```

```python
sns.stripplot(data=penguins, x = "species", y = "body_mass_g", hue="island", dodge=True, jitter=True)
```

```python
sns.set_context("talk")
sns.swarmplot(data=penguins, x = "species", y = "body_mass_g", hue="island")
sns.despine(left=True)
```

```python
sns.histplot(data=penguins, x = "body_mass_g")
```
